```python
from splinter import Browser
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pymongo
import requests
```

```python
browser = Browser('chrome')
```

```python
def get_soup(url):
    browser.visit(url)
    html = browser.html    
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
soup = get_soup("http://quotes.toscrape.com/page/1")
```

```python
quotes = soup.find_all('div', class_='quote')
```

```python
page_numbers = range(1,11)
quotes = []
for page in page_numbers:
    print(f"Scraping quotes info from {page}...")
    soup = get_soup(f"http://quotes.toscrape.com/page/{page}")
    quotes.append(soup.find_all('div', class_='quote'))
    
```
